```python
#EA,RA & ERA are highly negatively correleated with W so droping ERA &EA
df.drop('ER',axis=1,inplace=True)
df.drop('ERA',axis=1,inplace=True)
```

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
z
```

```python
df_new=df[(z<3).all(axis=1)] #removing outlier
```

```python
df_new
```

```python
df_new.shape
```

```python
x=df_new.iloc[:,1:16]
```

```python
x
```

```python
y=df_new.iloc[:,0].values
```

```python
y
```

```python
y=y.reshape(-1,1)
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
print("Pridicted result price:",pred)
print("Actual price:",y_test)
```

```python
print('Error')
print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean square error:',mean_squared_error(y_test,pred))
print('Root Mean square error:',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
#r2 score cofficient of determination
#is the proportion of the variance in the dependent variable that is pridictable from independent variable
#i.e change in y when ever there is change in x
```

```python
#Will reduce the cofficient to minimum
rd=Ridge()
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```
